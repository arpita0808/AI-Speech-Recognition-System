```python
# Import Whisper library for speech recognition
import whisper

# Import Javascript display functionality for microphone recording
from IPython.display import Javascript

# Import output module from Google Colab
from google.colab import output

# Import function to decode audio data
from base64 import b64decode

# Import scipy library for audio processing
import scipy.io.wavfile
```

```python
# Javascript code for recording audio from microphone
RECORD = """
const sleep = time => new Promise(resolve => setTimeout(resolve, time))

const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})

var record = async () => {

  // Request microphone access
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })

  // Create media recorder object
  recorder = new MediaRecorder(stream)

  // Store recorded audio chunks
  chunks = []

  // Save audio data chunks
  recorder.ondataavailable = e => chunks.push(e.data)

  // Start recording
  recorder.start()

  // Record audio for 15 seconds
  await sleep(15000)

  // Stop recording
  recorder.stop()

  // Wait until recording stops completely
  await new Promise(resolve => recorder.onstop = resolve)

  // Convert chunks into audio blob
  blob = new Blob(chunks)

  // Convert blob into text format
  text = await b2text(blob)

  // Return recorded audio
  return text
}
"""

# Display Javascript recording interface
display(Javascript(RECORD))

# Execute recording function
data = output.eval_js("record()")

# Decode recorded audio data
binary = b64decode(data.split(',')[1])

# Save recorded audio as WAV file
with open('audio.wav', 'wb') as f:
    f.write(binary)

# Print confirmation message
print("Recording saved as audio.wav")
```

```python
# Import torch library
import torch

# Check GPU availability
print("GPU Available:", torch.cuda.is_available())
```

```python
# Load pretrained Whisper model
model = whisper.load_model("medium")

# Print confirmation message
print("Whisper model loaded successfully")
```

```python
# Convert speech audio into text
result = model.transcribe("audio.wav")

# Print heading
print("Recognized Text:")

# Print transcription output
print(result["text"])
```

```python
# Detect language used in audio
print("Detected Language:")

# Print detected language
print(result["language"])
```

```python
# Generate transcription with timestamps
result = model.transcribe("audio.wav", word_timestamps=True)

# Loop through each speech segment
for segment in result['segments']:

    # Print start and end timestamps
    print(f"Start: {segment['start']} sec | End: {segment['end']} sec")

    # Print transcribed text for each segment
    print(segment['text'])
```

```python
from datetime import datetime

# Get current date and time
current_time = datetime.now()

# Open file in append mode
with open("transcription_output.txt", "a") as file:

    # Save timestamp
    file.write(f"\nTime: {current_time}\n")

    # Save detected language
    file.write(f"Language: {result['language']}\n")

    # Save transcription text
    file.write(f"Text: {result['text']}\n")

    # Add separator line
    file.write("-" * 50 + "\n")

# Print confirmation message
print("Transcription history updated successfully")
```
